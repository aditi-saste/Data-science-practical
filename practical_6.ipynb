np.NaN AND None

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({"rollno":[np.NaN,6205,6206,6207,6208],
                 "name":[np.NaN,"Aditi","perena","sanika","kalpana"],
                 "mname":["suresh","rajendra",np.NaN,"rajendra","jaganath"],
                 "lname":["patil","saste",np.NaN,"pawar",np.NaN],
                "class":["A","B","A","A","B"],"marks":[np.NaN,67,67,54,40]})
df
```

```python
df.isnull().sum()
```

```python
import pandas as pd
df=pd.DataFrame({"rollno":[np.NaN,6205,6206,6207,6208],
                 "Fname":["Aditi rajendra saste","suhani jaganath saste","perena sanjay shelke","sanika tanaji patil","mansi vijay jagdale"],
                "class":["A","B","A","A","B"],"marks":[np.NaN,67,67,54,40]})
df
```

data manupulation

```python
df["rollno"]=df["rollno"].fillna(6204)
df["rollno"]
```

```python
df
```

```python
mean_marks=df["marks"].mean()
mean_marks
```

```python
df["meanmarks"]=df["marks"].fillna(mean_marks)
df["meanmarks"]
```

```python
df
```

```python
median_marks=df["marks"].median()
median_marks
```

```python
df["meadian_marks"]=df["marks"].fillna(median_marks)
df["meadian_marks"]
```

```python
mode_marks=df["marks"].mode()
mode_marks
```

```python
df["modemarks"]=df["marks"].fillna(mode_marks)
df["modemarks"]
```

```python
df
```

```python
df["fname"]=df["Fname"].str.split().str[0]
df
```
